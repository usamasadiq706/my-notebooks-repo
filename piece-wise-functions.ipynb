# 1. Plot the following **Zero-Order** Piece-wise function:

$$
 \Large f(t) = \sum_{k = 1}^{4} \alpha_{k} . b_{o}(t-k)
$$
 
***where***

$$
  \large \alpha_{1} = 2, \alpha_{2} = 3, \alpha_{3} = 1, \alpha_{4} = -1 
$$

***and***

$$ \Large
b_{o}(t)=
\begin{cases}
1 & \quad -1/2 \le t \le 1/2 \\ 
0 & \quad \text{else}
\end{cases}
$$



```python
# Defining and Intiliazing Variable

res = 5.0   # Set the resolution of time-axis.
t = np.linspace(-res, +res, 1000) # Values of time-axis. 



# Define Base function using np.piecewise using conditions list & corresponding functions list using lambda functions. 

b0_t = np.piecewise(t, 
                    [t < -0.5, ((t >= -0.5) & (t <= 0.5)), t > 0.5],                     
                    [lambda x: 0, lambda x: 1, lambda x: 0]
                   )



# Deifne base function shifted to right by 1 unit along timea-axis i.e. b0_t(t-1). 

b0_t_1 = np.piecewise(t, 
                     [t < -0.5 + 1, ((t >= -0.5 + 1) & (t <= 0.5 + 1)), t > 0.5 + 1 ], 
                     [lambda x: 0, lambda x: 1, lambda x: 0]
                     )


# Deifne base function shifted to right by 2 unit along timea-axis i.e. b0_t(t-2). 

b0_t_2 = np.piecewise(t, 
                     [t < -0.5 + 2, ((t >= -0.5 + 2) & (t <= 0.5 + 2)), t > 0.5 + 2 ], 
                     [lambda x: 0, lambda x: 1, lambda x: 0]
                     )

# Deifne base function shifted to right by 3 unit along timea-axis i.e. b0_t(t-3). 

b0_t_3 = np.piecewise(t, 
                     [t < -0.5 + 3, ((t >= -0.5 + 3) & (t <= 0.5 + 3)), t > 0.5 + 3 ], 
                     [lambda x: 0, lambda x: 1, lambda x: 0]
                     )

# Deifne base function shifted to right by 4 unit along timea-axis i.e. b0_t(t-4). 

b0_t_4 = np.piecewise(t, 
                     [t < -0.5 + 4, ((t >= -0.5 + 4) & (t <= 0.5 + 4)), t > 0.5 + 4 ], 
                     [lambda x: 0, lambda x: 1, lambda x: 0]
                     )



'''
Define Main function as in equation 1 above.
'''

a_1, a_2, a_3, a_4 = 2, 3, 1, -1 # co-eficients

f_t = a_1* b0_t_1 + a_2 * b0_t_2 +  a_3 * b0_t_3 +  a_4 * b0_t_4 # function f(t) in equation above. 



##############################
# Plot Shifted Base Functions: 
##############################

fig, axs = plt.subplots(2, 2, figsize = (10,10)) # Subplots for four shifted base functions.


axs[0, 0].plot(t, b0_t, 'tab:blue')                # Plot base function bo(t) in blue color.
axs[0, 0].set_title(r'Base Function $\beta_o(t)$') # Set title of the plot in Mathtext,Matplotlib's internal LaTeX parser and layout engine.

axs[0, 1].plot(t, b0_t_1, 'tab:orange')            # Plot base function bo(t-1) in orange color.
axs[0, 1].set_title(r'Base Function $\beta_o(t-1)$')

axs[1, 0].plot(t, b0_t_2, 'tab:green')              # Plot base function bo(t-2) in green color.  
axs[1, 0].set_title(r'Base Function $\beta_o(t-2)$')

axs[1, 1].plot(t, b0_t_3, 'tab:red')                # Plot base function bo(t-3) in orange color.
axs[1, 1].set_title(r'Base Function $\beta_o(t-3)$')


# Setttings of axes of plots.
for i, ax in enumerate(axs.flat):
    
    y_labels = [r'$\beta_o(t)$', r'$\beta_o(t-1)$', r'$\beta_o(t-2)$', r'$\beta_o(t-3)$']
    ax.set(xlabel = 't', ylabel = y_labels[i]) # setting label of x and y-axis of subplots.
    
    # Grid settings for all subplots:
    ax.grid(True, which='major', color='#DDDDDD', linewidth=0.8) 
    ax.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
    ax.minorticks_on()
    
    ax.axhline(color = 'Black', linewidth = 1) # Show x-axis as bold black line in subplot.
    ax.axvline(color = 'Black', linewidth = 1) # Show y-axis as bold black line in subplot.



###########################
# Plot Piece-Wise Fucntion:
###########################

fig_n = plt.figure(figsize = (5,5)) # New figure for new single plot.
plt.plot(t, f_t, 'tab:purple' ) # plotting orignal funtion

plt.title(r'Zero Order Function $f(t)$') # set title of the plot.
plt.xlabel("t")                          # set x-axis label.
plt.ylabel(r'$f(t)$')                    # set y-axis label.  

# Grid Settings:
plt.grid(True, which='major', color='#DDDDDD', linewidth=0.8)
plt.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
plt.minorticks_on()

plt.axhline(color = 'Black', linewidth = 1)# Show x-axis as bold black line in subplot.
plt.axvline(color = 'Black', linewidth = 1)# Show yx-axis as bold black line in subplot.


plt.show() # show both figures.
```

# 2. Plot the following **First-Order** Piece-wise function:
$$
 \Large f(t) = \sum_{k = 1}^{4} \alpha_{k} . b_{1}(t-k)
$$
 
***where***

$$
  \large \alpha_{1} = 2, \alpha_{2} = 3, \alpha_{3} = 1, \alpha_{4} = -1 
$$

***and***

$$ \Large
b_{1}(t)=
\begin{cases}
1+t & \quad -1 \le t \le 0 \\ 
1-t & \quad 0 \le t \le 1 \\ 
0 & \quad \text{else}
\end{cases}
$$

```python
# Defining and Intiliazing Variable

res = 5.0   # Set the resolution of time-axis.
t = np.linspace(-res, +res, 1000) # Values of time-axis. 


# Define Base function using np.piecewise using conditions list & corresponding functions list using lambda functions. 

b1_t = np.piecewise(t, 
                    [t < -1.0, ((t >= -1.0) & (t <= 0.0)), ((t >= 0.0) & (t <= 1.0)),  t > 1.0], 
                    [lambda x: 0,  lambda x: 1+x, lambda x: 1-x, lambda x :0]
                   )

# Deifne base function shifted to right by 1 unit along timea-axis i.e. b1_t(t-1). 

b1_t_1 = np.piecewise(t, 
                     [t < -1.0 + 1.0, ((t >= -1.0 + 1.0) & (t <= 0.0 + 1)), ((t >= 0.0 + 1) & (t <= 1.0 + 1)),  t > 1.0 + 1 ], 
                     [lambda x: 0, lambda x: x, lambda x: 2-x, lambda x :0]
                     )


# Deifne base function shifted to right by 2 unit along timea-axis i.e. b1_t(t-2). 

b1_t_2 = np.piecewise(t, 
                     [t < -1.0 + 2, ((t >= -1.0 + 2) & (t <= 0.0 + 2)), ((t >= 0.0 + 2) & (t <= 1.0 + 2)),  t > 1.0 + 2 ],                     
                     [lambda x: 0, lambda x: x-1, lambda x: 3-x, lambda x :0]
                     )



# Deifne base function shifted to right by 3 unit along timea-axis i.e. b1_t(t-3). 

b1_t_3 = np.piecewise(t, 
                     [t < -1.0 + 3, ((t >= -1.0 + 3) & (t <= 0.0 + 3)), ((t >= 0.0 + 3) & (t <= 1.0 + 3)),  t > 1.0 + 3 ],                    
                     [lambda x: 0, lambda x: x-2, lambda x: 4-x, lambda x :0]
                     )


# Deifne base function shifted to right by 4 unit along timea-axis i.e. b1_t(t-4). 

b1_t_4 = np.piecewise(t, 
                     [t < -1.0 + 4, ((t >= -1.0 + 4) & (t <= 0.0 + 4)), ((t >= 0.0 + 4) & (t <= 1.0 + 4)),  t > 1.0 + 4 ],                     
                     [lambda x: 0, lambda x: x-3, lambda x: 5-x, lambda x :0]
                     )




'''
Define Main function as in equation 1 above.
'''
a_1, a_2, a_3, a_4 = 2, 3, 1, -1 # co-eficients

f_t = a_1* b1_t_1 + a_2 * b1_t_2 +  a_3 * b1_t_3 +  a_4 * b1_t_4 # function f(t) in equation above. 



##############################
# Plot Shifted Base Functions: 
##############################

fig, axs = plt.subplots(2, 2, figsize = (10,10)) # Subplots for four shifted base functions.


axs[0, 0].plot(t, b1_t, 'tab:blue')                # Plot base function bo(t) in blue color.
axs[0, 0].set_title(r'Base Function $\beta_1(t)$') # Set title of the plot in Mathtext,Matplotlib's internal LaTeX parser and layout engine.

axs[0, 1].plot(t, b1_t_1, 'tab:orange')            # Plot base function bo(t-1) in orange color.
axs[0, 1].set_title(r'Base Function $\beta_1(t-1)$')

axs[1, 0].plot(t, b1_t_2, 'tab:green')              # Plot base function bo(t-2) in green color.  
axs[1, 0].set_title(r'Base Function $\beta_1(t-2)$')

axs[1, 1].plot(t, b1_t_3, 'tab:red')                # Plot base function bo(t-3) in orange color.
axs[1, 1].set_title(r'Base Function $\beta_1(t-3)$')


# Setttings of axes of plots.
for i, ax in enumerate(axs.flat):
    
    y_labels = [r'$\beta_1(t)$', r'$\beta_1(t-1)$', r'$\beta_1(t-2)$', r'$\beta_1(t-3)$']
    ax.set(xlabel = 't', ylabel = y_labels[i]) # setting label of x and y-axis of subplots.
    
    # Grid settings for all subplots:
    ax.grid(True, which='major', color='#DDDDDD', linewidth=0.8) 
    ax.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
    ax.minorticks_on()
    
    ax.axhline(color = 'Black', linewidth = 1) # Show x-axis as bold black line in subplot.
    ax.axvline(color = 'Black', linewidth = 1) # Show y-axis as bold black line in subplot.



###########################
# Plot Piece-Wise Fucntion:
###########################

fig_n = plt.figure(figsize = (5,5)) # New figure for new single plot.
plt.plot(t, f_t, 'tab:purple' ) # plotting orignal funtion

plt.title(r'First Order Function $f(t)$') # set title of the plot.
plt.xlabel("t")                          # set x-axis label.
plt.ylabel(r'$f(t)$')                    # set y-axis label.  

# Grid Settings:
plt.grid(True, which='major', color='#DDDDDD', linewidth=0.8)
plt.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
plt.minorticks_on()

plt.axhline(color = 'Black', linewidth = 1)# Show x-axis as bold black line in subplot.
plt.axvline(color = 'Black', linewidth = 1)# Show yx-axis as bold black line in subplot.


plt.show() # show both figures.

```

# 3. Plot the following piece-wise _approximation_ of Sinc function:
$$
 \Large f(t) = \sum_{k = 1}^{4} \alpha_{k} . b_{2}(t-k)
$$
 
***where***

$$
  \large \alpha_{1} = 2, \alpha_{2} = 3, \alpha_{3} = 1, \alpha_{4} = -1 
$$

***and*** 

$$ \Large
b_{2}(t)=
\begin{cases}
\frac{(t + \frac{3}{2})^2}{2} & \quad -\frac{3}{2} \le t \le -\frac{1}{2} \\ 
-t^2 + \frac{3}{4} & \quad -\frac{1}{2} \le t \le \frac{1}{2} \\ 
\frac{(t - \frac{3}{2})^2}{2} & \quad \frac{1}{2} \le t \le \frac{3}{2} \\ 
0 & \quad \text{else}
\end{cases}
$$


```python
# Defining and Intiliazing Variable

res = 8.0   # Set the resolution of time-axis.
t = np.linspace(-res, +res, 1000) # Values of time-axis. 


# Define Base function using np.piecewise using conditions list & corresponding functions list using lambda functions. 

b2_t = np.piecewise(t, 
                    [t < -3/2, ((t >= -3/2) & (t <= -1/2)), ((t >= -1/2) & (t <= 1/2)), ((t >= 1/2) & (t <= 3/2)),  t > 3/2], 
                    [lambda x: 0,  lambda x: ((x + 3/2)**2)/2, lambda x:-(x**2) + 3/4 ,lambda x: ((x - 3/2)**2)/2,  lambda x :0]
                   )

# Deifne base function shifted to right by 1 unit along timea-axis i.e. b2_t(t-1). 

b2_t_1= np.piecewise(t, 
                    [t < -3/2 + 1, ((t >= -3/2 + 1) & (t <= -1/2 + 1)), ((t >= -1/2 + 1) & (t <= 1/2 + 1)), ((t >= 1/2 + 1) & (t <= 3/2 + 1)),  t > 3/2 + 1], 
                    [lambda x: 0,  lambda x: ((x-1 + 3/2)**2)/2, lambda x:-((x-1)**2) + 3/4 ,lambda x: (((x-1) - 3/2)**2)/2,  lambda x :0]
                    )


# Deifne base function shifted to right by 2 unit along timea-axis i.e. b2_t(t-2). 

b2_t_2= np.piecewise(t, 
                    [t < -3/2 + 2, ((t >= -3/2 + 2) & (t <= -1/2 + 2)), ((t >= -1/2 + 2) & (t <= 1/2 + 2)), ((t >= 1/2 + 2) & (t <= 3/2 + 2)),  t > 3/2 + 2], 
                    [lambda x: 0,  lambda x: ((x-2 + 3/2)**2)/2, lambda x:-((x-2)**2) + 3/4 ,lambda x: (((x-2) - 3/2)**2)/2,  lambda x :0]
                    )




# Deifne base function shifted to right by 3 unit along timea-axis i.e. b2_t(t-3). 

b2_t_3= np.piecewise(t, 
                    [t < -3/2 + 3, ((t >= -3/2 + 3) & (t <= -1/2 + 3)), ((t >= -1/2 + 3) & (t <= 1/2 + 3)), ((t >= 1/2 + 3) & (t <= 3/2 + 3)),  t > 3/2 + 3], 
                    [lambda x: 0,  lambda x: ((x-3 + 3/2)**2)/2, lambda x:-((x-3)**2) + 3/4 ,lambda x: (((x-3) - 3/2)**2)/2,  lambda x :0]
                    )


# Deifne base function shifted to right by 4 unit along timea-axis i.e. b2_t(t-4). 

b2_t_4= np.piecewise(t, 
                    [t < -3/2 + 4, ((t >= -3/2 + 4) & (t <= -1/2 + 4)), ((t >= -1/2 + 4) & (t <= 1/2 + 4)), ((t >= 1/2 + 4) & (t <= 3/2 + 4)),  t > 3/2 + 4], 
                    [lambda x: 0,  lambda x: ((x-4 + 3/2)**2)/2, lambda x:-((x-4)**2) + 3/4 ,lambda x: (((x-4) - 3/2)**2)/2,  lambda x :0]
                    )




'''
Define Main function as in equation 1 above.
'''
a_1, a_2, a_3, a_4 = 2, 3, 1, -1 # co-eficients (By changing theses we can chnage the final function f(t)).

f_t = a_1* b2_t_1 + a_2 * b2_t_2 +  a_3 * b2_t_3 +  a_4 * b2_t_4 # function f(t) in equation 3. 



##############################
# Plot Shifted Base Functions: 
##############################

fig, axs = plt.subplots(2, 2, figsize = (10,10)) # Subplots for four shifted base functions.


axs[0, 0].plot(t, b2_t, 'tab:blue')                # Plot base function bo(t) in blue color.
axs[0, 0].set_title(r'Base Function $\beta_2(t)$') # Set title of the plot in Mathtext,Matplotlib's internal LaTeX parser and layout engine.

axs[0, 1].plot(t, b2_t_1, 'tab:orange')            # Plot base function bo(t-1) in orange color.
axs[0, 1].set_title(r'Base Function $\beta_2(t-1)$')

axs[1, 0].plot(t, b2_t_2, 'tab:green')              # Plot base function bo(t-2) in green color.  
axs[1, 0].set_title(r'Base Function $\beta_2(t-2)$')

axs[1, 1].plot(t, b2_t_3, 'tab:red')                # Plot base function bo(t-3) in orange color.
axs[1, 1].set_title(r'Base Function $\beta_3(t-3)$')


# Setttings of axes of plots.
for i, ax in enumerate(axs.flat):
    
    y_labels = [r'$\beta_2(t)$', r'$\beta_2(t-1)$', r'$\beta_2(t-2)$', r'$\beta_2(t-3)$']
    ax.set(xlabel = 't', ylabel = y_labels[i]) # setting label of x and y-axis of subplots.
    
    # Grid settings for all subplots:
    ax.grid(True, which='major', color='#DDDDDD', linewidth=0.8) 
    ax.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
    ax.minorticks_on()
    
    ax.axhline(color = 'Black', linewidth = 1) # Show x-axis as bold black line in subplot.
    ax.axvline(color = 'Black', linewidth = 1) # Show y-axis as bold black line in subplot.



###########################
# Plot Piece-Wise Fucntion:
###########################

fig_n = plt.figure(figsize = (5,5)) # New figure for new single plot.
plt.plot(t, f_t, 'tab:purple' ) # plotting orignal funtion

plt.title(r'Piece-sie Sinc $f(t)$') # set title of the plot.
plt.xlabel("t")                          # set x-axis label.
plt.ylabel(r'$f(t)$')                    # set y-axis label.  

# Grid Settings:
plt.grid(True, which='major', color='#DDDDDD', linewidth=0.8)
plt.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
plt.minorticks_on()

plt.axhline(color = 'Black', linewidth = 1)# Show x-axis as bold black line in subplot.
plt.axvline(color = 'Black', linewidth = 1)# Show yx-axis as bold black line in subplot.


plt.show() # show both figures.
```

# 5. Plot the following piece-wise Bump function:
$$
 \Large f(t) = \sum_{k = 1}^{4} \alpha_{k} . b_{3}(t-k)
$$
 
***where***

$$
  \large \alpha_{1} = 2, \alpha_{2} = 3, \alpha_{3} = 1, \alpha_{4} = -1 
$$

***and*** 

$$ \Large
b_{4}(t)=
\begin{cases}
\exp -(\frac{1}{1 - x^2}) & \quad -1 < x < 1 \\ 
0 & \quad else
\end{cases}
$$


```python
# Defining and Intiliazing Variable

res = 10.0   # Set the resolution of time-axis.
t = np.linspace(-res, +res, 1000) # Values of time-axis. 


w = 10 # width of the bump. 


# Define Base function using np.piecewise using conditions list & corresponding functions list using lambda functions. 

b5_t = np.piecewise(t, 
                    [t <= -w/2, ((t > -w/2) & (t < +w/2)), t >= +w/2], 
                    [lambda x: 0,  lambda x: np.exp(-1/((w/2)**2-x**2)),  lambda x :0]
                   )

# Deifne base function shifted to right by 1 unit along timea-axis i.e. b5_t(t-1). 

b5_t_1 = np.piecewise(t, 
                    [t <= -w/2 + 1, ((t > -w/2 + 1) & (t < +w/2 + 1)), t >= +w/2 +1], 
                    [lambda x: 0,  lambda x: np.exp(-1/((w/2)**2-(x-1)**2)),  lambda x :0]
                   )

# Deifne base function shifted to right by 2 unit along timea-axis i.e. b5_t(t-2). 

b5_t_2 = np.piecewise(t, 
                    [t <= -w/2 + 2, ((t > -w/2 + 2) & (t < +w/2 + 2)), t >= +w/2 + 2], 
                    [lambda x: 0,  lambda x: np.exp(-1/((w/2)**2-(x-2)**2)),  lambda x :0]
                   )



# Deifne base function shifted to right by 3 unit along timea-axis i.e. b5_t(t-3). 

b5_t_3 = np.piecewise(t, 
                    [t <= -w/2 + 3, ((t > -w/2 + 3) & (t < +w/2 + 3)), t >= +w/2 + 3], 
                    [lambda x: 0,  lambda x: np.exp(-1/((w/2)**2-(x-3)**2)),  lambda x :0]
                   )

# Deifne base function shifted to right by 4 unit along timea-axis i.e. b3_t(t-4). 

b5_t_4 = np.piecewise(t, 
                    [t <= -w/2 + 4, ((t > -w/2 + 4) & (t < +w/2 + 4)), t >= +w/2 + 4], 
                    [lambda x: 0,  lambda x: np.exp(-1/((w/2)**2-(x-4)**2)),  lambda x :0]
                   )



'''
Define Main function as in equation 1 above.
'''
a_1, a_2, a_3, a_4 = 2, 3, 1, -1 # co-eficients (By changing theses we can chnage the final function f(t)).

f_t = a_1 * b5_t_1 + a_2 * b5_t_2 +  a_3 * b5_t_3 +  a_4 * b5_t_4 # function f(t) in equation 3. 




##############################
# Plot Shifted Base Functions: 
##############################

fig, axs = plt.subplots(2, 2, figsize = (10,10)) # Subplots for four shifted base functions.


axs[0, 0].plot(t, b5_t, 'tab:blue')                # Plot base function bo(t) in blue color.
axs[0, 0].set_title(r'Base Function $\beta_5(t)$') # Set title of the plot in Mathtext,Matplotlib's internal LaTeX parser and layout engine.

axs[0, 1].plot(t, b5_t_1, 'tab:orange')            # Plot base function bo(t-1) in orange color.
axs[0, 1].set_title(r'Base Function $\beta_5(t-1)$')

axs[1, 0].plot(t, b5_t_2, 'tab:green')              # Plot base function bo(t-2) in green color.  
axs[1, 0].set_title(r'Base Function $\beta_5(t-2)$')

axs[1, 1].plot(t, b5_t_3, 'tab:red')                # Plot base function bo(t-3) in orange color.
axs[1, 1].set_title(r'Base Function $\beta_5(t-3)$')


# Setttings of axes of plots.
for i, ax in enumerate(axs.flat):
    
    y_labels = [r'$\beta_5(t)$', r'$\beta_5(t-1)$', r'$\beta_5(t-2)$', r'$\beta_5(t-3)$']
    ax.set(xlabel = 't', ylabel = y_labels[i]) # setting label of x and y-axis of subplots.
    
    # Grid settings for all subplots:
    ax.grid(True, which='major', color='#DDDDDD', linewidth=0.8) 
    ax.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
    ax.minorticks_on()
    
    ax.axhline(color = 'Black', linewidth = 1) # Show x-axis as bold black line in subplot.
    ax.axvline(color = 'Black', linewidth = 1) # Show y-axis as bold black line in subplot.



###########################
# Plot Piece-Wise Fucntion:
###########################

fig_n = plt.figure(figsize = (5,5)) # New figure for new single plot.
plt.plot(t, f_t, 'tab:purple' ) # plotting orignal funtion

plt.title(r'Piece-wise Bump $f(t)$') # set title of the plot.
plt.xlabel("t")                          # set x-axis label.
plt.ylabel(r'$f(t)$')                    # set y-axis label.  

# Grid Settings:
plt.grid(True, which='major', color='#DDDDDD', linewidth=0.8)
plt.grid(True, which='minor', color='#EEEEEE', linestyle=':', linewidth=0.5)
plt.minorticks_on()

plt.axhline(color = 'Black', linewidth = 1)# Show x-axis as bold black line in subplot.
plt.axvline(color = 'Black', linewidth = 1)# Show yx-axis as bold black line in subplot.


plt.show() # show both figures.
```
